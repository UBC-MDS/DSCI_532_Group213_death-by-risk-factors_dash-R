```r
library(tidyverse)
library(dplyr)
library(ggplot2)
```

```r
factors_data=read.csv("clean_data_line_plot.csv")
factors_data$obesity <- factors_data$obesity / 1000000
```

```r
line_plot <- ggplot(factors_data, aes(x= year, y = obesity, color=continent)) +
             geom_line(size=0.5)+
             geom_point()+
             theme(axis.text.x = element_text(angle = -60), aspect.ratio = 0.5)+
             scale_x_continuous(breaks = seq(1990, 2020, 1))+
             xlab("Years") +
             ylab("Number of Deaths (millions)")+
             ggtitle("Trend of Obesity Deaths over the years in continents")
line_plot
```
